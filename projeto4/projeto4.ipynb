```python
image = os.path.join(os.getcwd(), "RPS/snap_00000.png")
image = skimage.io.imread(image, as_gray=True)

def ski_plot(img):

    plt.figure(figsize=(10, 10))
    plt.imshow(img, cmap='gray', interpolation='nearest')
    plt.axis('off')
    plt.show()

print(image.shape)    
image = image[100:400,100:600]
ski_plot(image)
```

```python

from skimage import io
from skimage.color import rgb2hsv

play = rgb2hsv(io.imread("RPS/pedra_0.png"))
plt.figure()
plt.imshow(play[:,:,0], cmap="gray")
plt.figure()
plt.imshow(play[:,:,1], cmap="gray")
plt.figure()
plt.imshow(play[:,:,2], cmap="gray")
img = play[:,:,1]
plt.show()
```

```python
play = rgb2hsv(io.imread("RPS/pedra_1.png"))
img = play[:,:,1]
img_cut = img[100:400,150:500]
img_the = (img_cut>0.4) & (img_cut<1.1)
ski_plot(img_the)
```

```python
x,y = img_the.shape
reco = np.zeros(img_the.shape)
reco[(x//2),(y//2)] =1
```

```python
print(reco.max())
```

```python

a = skimage.morphology.reconstruction(reco,img_the.astype(np.uint8))
```

```python
ski_plot(a)
```
